### Exercise Question 8: 
Calculate total sale data for last year for each product and show it using a Pie chart

```python
color=['blue', 'orange', 'pink', 'green', 'purple', 'c']
plt.figure(figsize=(10, 6))
categories = data.columns[1:7]
sales_data = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(),
              data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum()]
plt.pie(sales_data, labels=categories, autopct='%1.2f%%',colors=color)
plt.legend(loc='lower right')
plt.show()
```

### Exercise Question 10: 
Read all product sales data and show it using the stack plot

```python
color=['blue', 'orange', 'pink', 'green', 'purple', 'c']
cream_sales = data['facecream'].tolist()
wash_sales = data['facewash'].tolist()
toothpaste_sales = data['toothpaste'].tolist()
bathingsoap_sales = data['bathingsoap'].tolist()
shampoo_sales = data['shampoo'].tolist()
moisturizer_sales = data['moisturizer'].tolist()
labels=["monthList","cream_sales", "wash_sales", "toothpaste_sales", "bathingsoap_sales",
              "shampoo_sales", "moisturizer_sales"]
plt.stackplot(monthList, cream_sales, wash_sales, toothpaste_sales, bathingsoap_sales,
              shampoo_sales, moisturizer_sales, colors=color,labels=labels)
plt.legend(loc='lower right')
plt.xlim(1,13)
plt.show()
```

#Analysis
Scatterplot displays the complete data for visualization. It shows each part stacked onto one another and how each part makes the complete figure.
Shampoo sales are on top of stack plot followed by bathing soap then toothpaste then wash sales then cream sales.
Highest sale made by shampoo was in month of december

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# pie chart presenting the male/female proportion
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic.head()
count = titanic['sex'].value_counts()
count.values.tolist() # male-843,female-466 hence op-[843,466]
count.index.values.tolist()
colors = ['blue', 'orange']
f, (ax1, ax2) = plt.subplots(1,2)#creat subplots with 1 Row And 2 Col, Create figure
ax1.pie(count.values.tolist(), labels = count.index.values.tolist(), colors=colors, startangle=90, autopct='%.1f%%')
ax1.set_title('Male/Female proportion')

# scatter plot of fare and age with c as sex data

titanic = titanic.dropna(subset = ["sex"]) # as there is nan value, have to drop it 
colors = {'male':'blue', 'female':'orange'}
ax2.scatter(titanic.age,titanic.fare,c = titanic['sex'].iloc[:1310].apply(lambda x: colors[x]))
ax2.set_xlabel('Age')
ax2.set_ylabel('Fare')
ax2.set_title('Fare paid and the Age')
plt.show()
```
