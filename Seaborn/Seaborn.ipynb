# 12. Find the normal distribution, binomial distribution, poison distribution for the randomly generated 50 values and visualize it using distance plot.

```python
from numpy import random
from pylab import *
sns.set(rc={"figure.figsize": (10, 10)});
subplot(3,1,1)
binomial_data=np.random.binomial(n=10,p=0.3,size=1000)
sns.distplot(binomial_data,kde=True,hist=True,color="red")
poison_data=np.random.binomial(n=10,p=0.3,size=1000)
subplot(3,1,2)
sns.distplot(poison_data,kde=True,hist=True,color="green")
normal_data=np.random.binomial(n=10,p=0.3,size=1000)
subplot(3,1,3)
sns.distplot(normal_data,kde=True,hist=True,color="purple")
```

#15. Generate the following 3 Plots as an output by setting the suitable parameters in countplot.

```python

import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)
sns.countplot(data=data, x="Payment", order = data['Payment'].value_counts().index[::-1],hue='Branch')
plt.title("Payment channel by branch")
plt.subplot(1, 2, 2)
plt.title("Gender_count")
sns.countplot(data=data, x="Gender", order = data['Gender'].value_counts().index[::-1])
plt.show()
plt.subplot(1, 1, 1)
plt.title("Payment per product Line")
sns.countplot(data=data, y="Product line", order = data['Product line'].value_counts().index[::-1],hue='Payment')
plt.show()
```
